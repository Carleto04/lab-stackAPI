# StackAPI

#### Import the necessary libraries here:

```python
import time
import datetime
import json
import requests
import pandas as pd
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
```

```python
# Time stamp is a way to track time as a running total of seconds.
# This count starts at the Unix Epoch on January 1st, 1970.

date1 = '07/06/2020'
date2 = '07/07/2020'
ts1 = time.mktime(time.strptime(date1, "%d/%m/%Y"))
ts2 = time.mktime(time.strptime(date2, "%d/%m/%Y"))
```

```python
print(ts1)
print(ts2)
```

#### Question 1: Find the questions and answers of last month.

```python
Q = SITE.fetch('questions', fromdate=1591480800, todate=1594072800) 
#import the data with filters 'Date range' and category 'questions'
A = SITE.fetch('answers', fromdate=1591480800, todate=1594072800)
#import the data with filters 'Date range' and category 'answers'
```

```python
from pandas.io.json import json_normalize
q_df = pd.json_normalize(Q['items'])
a_df = pd.json_normalize(A['items'])

#create a DataFrame with the column item
```

#### Question 2: Find the most voted question from last month till today, with at least a score of 5 and tagged with 'python'. 

```python
QTOP = SITE.fetch('questions', fromdate=1591480800, todate=1594072800, min=5, tagged='python', sort='votes')
QTOP_df = pd.json_normalize(QTOP['items'])
TOPquestion = QTOP_df['question_id'][0]
```
